```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0,1,-1,2,-2])
f = np.array([-5,-3,-15,39,-9])
```

```python
eva_points = np.zeros_like([x]*5)

for i in range(len(x)):
    num = x[i]
 
    #Â basic functions
    temp = 1
    for j in range(5):
        eva_points[i][j] = temp
        temp *= num

eva_points
```

```python
s = np.linalg.inv(eva_points) @ f
s
```

```python
big_x = np.linspace(min(x), max(x), 200)
big_y = []

for n in big_x:
    val = 0
    for i in range(len(x)):
        coef = s[i]
        for j in range(i):
            coef *= n
        val += coef
    big_y.append(val)

```

```python
plt.plot(big_x, big_y)
plt.scatter(x,f, color="red")
plt.show()
```
