```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
def lagrance_interpolation(point, x_list, y_list):
    approximate = 0
    for i in range(len(x_list)):
        top = 1
        low = 1
        for j in range(len(x_list)):
            if(i != j):
                top *= (point - x_list[j])
                low *= (x_list[i] - x_list[j]) 
        approximate += y_list[i] * (top / low)

    return approximate
```

```python
x_list = [1, 2, 3, 5, 6]
y_list = [4.75, 4, 5.25, 19.75, 36]

x_list_used = [2, 3, 5, 6]
y_list_used = [4, 5.25, 19.75, 36]

result = lagrance_interpolation(4, x_list_used, y_list_used)
print(result)
```

```python
xx = np.arange(min(x_list)-1, max(x_list)+1, 0.0001)
yy = list()

start = time.time()
for x in xx:
    yy.append(lagrance_interpolation(x, x_list, y_list))
end = time.time()
print("Time:", end - start)

plt.scatter(x_list, y_list, color="red", label="real data")
plt.plot(xx, yy, label="interpolation")
plt.legend()
plt.show()
```
