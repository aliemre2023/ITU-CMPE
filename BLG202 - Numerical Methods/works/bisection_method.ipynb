```python
import time
import matplotlib.pyplot as plt
from math import exp
```

```python
interval = [0, 1]
equation = lambda x: exp(-x) - x
```

```python
def result(eq, inter):
    x1 = inter[0]
    x2 = inter[1]
    mid = (x1 + x2) / 2

    return eq(x1),x1, eq(x2),x2 , eq(mid),mid
```

```python
x_vals = []
time_vals = []
loop = 0
```

```python
start_time = time.time()
while(True):
    loop += 1
    current_time = time.time() - start_time
    r1,x1 , r2,x2 , rmid,mid = result(equation, interval)
    x_vals.append(mid)
    time_vals.append(current_time)
    if(r1 * r2 < 0):
        if(r1 * rmid < 0):
            interval = [x1, mid]
        else:
            interval = [mid, x2]
    else:
        print("No root")
        break
    
    if(abs(x1 - x2) < 1e-3):
        break
end_time = time.time()

print(f"Root: {((interval[0]+interval[1])/2):.9f}")
print(f"Runtime: {end_time - start_time}")
```

```python
plt.plot(x_vals, time_vals ,marker="o")
plt.ylim(min(time_vals), max(time_vals))

plt.xlabel('X Values')
plt.ylabel('Time')
plt.title(f'Bisection Method, Loop : {loop}')

plt.show()
```
