```python
import time
import matplotlib.pyplot as plt
from math import exp
from sympy import symbols, exp, diff
```

```python
start_x = 1
equation = lambda x: exp(-x) - x
```

```python
x = symbols('x')
derivative = diff(equation(x), x)
def result(eq, curr_x):
    outp_1 = eq(curr_x)
    outp_2 = derivative.subs(x, curr_x)
    next_x = curr_x - (outp_1 / outp_2)
    return next_x
```

```python
x_vals = []
time_vals = []
loop = 0
```

```python
start_time = time.time()
while(True):
    loop += 1
    current_time = time.time() - start_time
    x_vals.append(start_x)
    next_x = result(equation, start_x)
    time_vals.append(current_time)
    
    if(abs(start_x - next_x) < 1e-3):
        start_x = next_x
        break
    start_x = next_x
end_time = time.time()

print(f"Root: {start_x:.9f}")
print(f"Runtime: {end_time - start_time}")
```

```python
plt.plot(x_vals, time_vals, marker="o")
plt.ylim(min(time_vals), max(time_vals))

plt.xlabel('X Values')
plt.ylabel('Time')
plt.title(f'Newton Method, Loop: {loop}')

plt.show()
```
