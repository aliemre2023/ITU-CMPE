```python
import time
import matplotlib.pyplot as plt
from math import exp
```

```python
storage_x = [0, 1]
equation = lambda x: exp(-x) - x
```

```python
def result(eq, storage_x):
    x1 = storage_x[0]
    x2 = storage_x[1]

    y1 = eq(x1)
    y2 = eq(x2)
    
    if((x2 - x1) == 0):
        return x1

    x_next = x2 - (y2 / ((y2 - y1) / (x2 - x1)))
    
    storage_x = [x2, x_next]

    return storage_x
```

```python
x_vals = []
time_vals = []
loop = 0
```

```python
start_time = time.time()
while(True):
    loop += 1
    current_time = time.time() - start_time
    storage_x = result(equation, storage_x)
    x_vals.append(storage_x[1])
    time_vals.append(current_time)
    
    if(abs(storage_x[0] - storage_x[1]) < 1e-3):
        break

end_time = time.time()

print(f"Root: {(storage_x[1]):.9f}")
print(f"Runtime: {end_time - start_time}")
```

```python
plt.plot(x_vals, time_vals, marker="o")
plt.ylim(min(time_vals), max(time_vals))

plt.xlabel('X Values')
plt.ylabel('Time')
plt.title(f'Secant Method, Loop: {loop}')

plt.show()
```
