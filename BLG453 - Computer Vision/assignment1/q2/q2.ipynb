```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("landscape.png")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
def gamma_correction(image, gamma):
    gamma_corrected = np.power(image / 255.0, gamma)
    gamma_corrected = np.uint8(gamma_corrected * 255)

    plt.figure(figsize=(6,6))
    plt.imshow(gamma_corrected)
    plt.axis('off')
    plt.show()

    return gamma_corrected
```

```python
gamma01 = gamma_correction(image_rgb, 0.1)
```

```python
gamma04 = gamma_correction(image_rgb, 0.4)
```

```python
gamma10 =gamma_correction(image_rgb, 1)
```

```python
gamma15 = gamma_correction(image_rgb, 1.5)
```

```python
gamma22 = gamma_correction(image_rgb, 2.2)
```

```python
gammas = [0.1, 0.4, 1, 1.5, 2.2]
x = np.arange(256)

plt.figure(figsize=(8,6))
for g in gammas:
    y = ((x / 255.0) ** g) * 255
    plt.plot(x, y, label=f'gamma={g}')
plt.title('Gamma transformation curves')
plt.xlabel('Input intensity (0-255)')
plt.ylabel('Output intensity (0-255)')
plt.xlim(0,255)
plt.ylim(0,255)
plt.legend()
plt.grid(True)
plt.show()
```

# Results

gamma = 1  ==> same image <br>
gamma < 1  ==> makes the image brighter <br>
gamma > 1  ==> makes the image darker <br>

As the transformation curve shows, while gamma is smaller than 1, the values which is coles to the 0 will be closer to the 255 after the gamma correction, visa versa for gamma > 1.

This can be understandable from the gamma_correction formula.



(a) Which gamma value provides the best detail enhancement?
-> 0.4 looks like good sunset

(b) Why is gamma correction more effective than simply increasing pixel intensity?
-> Linear increment directly brighten everything. In the gamma correction, highlights, shadows process more effectively, like bending the tones
