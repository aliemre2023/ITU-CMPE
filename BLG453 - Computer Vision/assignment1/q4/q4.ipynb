```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "findme.png"

image = cv2.imread(image_path)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
img_int = image_rgb.astype(np.int32)
R = img_int[:, :, 0]
G = img_int[:, :, 1]
B = img_int[:, :, 2]
```

```python
new_red = -(R + G + B)

processed = image_rgb.copy()
processed[:, :, 0] = new_red

image_rgb = processed
```

```python
plt.figure(figsize=(6,6))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
